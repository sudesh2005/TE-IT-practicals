```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('bank_data_set.csv')
data.head(5)
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.distplot(data.age[data.deposit=='yes'],color='red',label='deposit=yes')
plt.legend()
plt.show()
```

```python
sns.distplot(data.age[data.deposit=='no'],color='blue',label='deposit=no')
plt.legend()
plt.show()
```

```python
#Inference : The people who are in age gap 24 to 50 are more likely to buy a term deposit
```

```python
plt.figure(figsize=(19, 6))
sns.countplot(x='age', hue = 'marital', data = data)
plt.title("Age VS Count")
```

```python
sns.boxplot(x='marital', y='age', data=data)
plt.title('Boxplot of Age by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Age')
plt.show()
```

```python
sns.countplot(x='marital', hue ='deposit', data = data)
plt.title('Count vs. Deposit')
```

```python
#Inference : From the above 3 graph we get that most number of count is in age gap 24 - 40.
#Among these the highest number of count is of married people who are more likely to buy term deposit from other two.
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='deposit', y='balance', data=data)
plt.title('Violin Plot of Balance vs. Deposit')
plt.xlabel('Deposit')
plt.ylabel('Balance')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.distplot(data.balance[data.deposit=='yes'], color='green', label='deposit=yes')
plt.legend()
plt.title('Age vs Balance Distribution')
plt.xlabel('Balance')
plt.ylabel('Density')
plt.show()
```

```python
#inference : The people with balance less than 10,000 are more likely to buy term deposit
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='contact', hue='deposit', data=data)
plt.title('Contact Method vs. Deposit')
plt.xlabel('Contact Method')
plt.ylabel('Count')
plt.show()


```

```python
#Inference : The Cellular method is widely used for contacting and is the most effective among th rest.
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data[data['deposit'] == 'yes']['duration'],  color='green', label='Deposit=yes')
sns.histplot(data[data['deposit'] == 'no']['duration'],  color='red', label='Deposit=no')
plt.title('Duration vs. Deposit')
plt.xlabel('Duration of Contact (seconds)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#Inference : The most effective duration is 250 to 1100 for calling
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='month', hue='deposit', data=data)
plt.title('Deposit by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
#Inference : In May month term deposit rate is highest , while in December and march it is lowest.
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='job', hue='deposit', data=data)
plt.title('Job vs. Deposit')
plt.xlabel('Job')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Inference : The people working in Management sector have the highest deposit rate while housemaid and entrepreneur have lowest rate.
```

```python
plt.figure(figsize=(12, 6))
dep_no = data[data['deposit'] == 'no']
sns.countplot(x='month', hue='contact', data= dep_no)
plt.title('Contact Method Rejected most')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
#inference : Cellular contact done in August , July and May results in rejection. In May and June the contect from unkown gets most rejection.
```
